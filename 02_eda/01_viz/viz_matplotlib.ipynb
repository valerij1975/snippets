# Plot formatting

## style

```python
list_style=plt.style.available
style = widgets.Dropdown(options=list_style, value='seaborn-v0_8')
def change_style(choose_style):
    print('ok')
    plt.style.use(choose_style)
    fig = df_tge['fixing_1_kurs'].plot(figsize=(15, 10));
    fig.set_title("Fixing kurs of energy")
    fig.set_xlabel("Date-Hour")
    fig.set_ylabel("PLN");
    
out = widgets.interactive_output(change_style, {'choose_style': style})
display(style, out)
```
